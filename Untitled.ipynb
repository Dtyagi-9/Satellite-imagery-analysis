```python
import rasterio
from rasterio import plot
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
import os
os.listdir('./photos')
```

```python
band4 = rasterio.open('./photos/LC82010222017097LGN00_B4.TIF')
band5 = rasterio.open('./photos/LC82010222017097LGN00_B5.TIF') 
```

```python
plot.show(band4)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
plot.show(band4, ax=ax1, cmap='Blues') #red
plot.show(band5, ax=ax2, cmap='Blues') #nir
fig.tight_layout()
```

```python
red = band4.read(1).astype('float64')
nir = band5.read(1).astype('float64')

nir
```

```python
np.seterr(divide='ignore', invalid='ignore')
ndvi=np.where(
    (nir+red)==0., 
    0, 
    (nir-red)/(nir+red))
ndvi
```

```python
ndviImage = rasterio.open('./sih-data/ndvi.tif','w',driver='Gtiff',
                          width=band4.width, 
                          height = band4.height, 
                          count=1, crs=band4.crs, 
                          transform=band4.transform, 
                          dtype='float64')
ndviImage.write(ndvi,1)
ndviImage.close()
```

```python
ndvi = rasterio.open('./sih-data/ndvi.tif')
fig = plt.figure(figsize=(18,12))
plot.show(ndvi)
```
